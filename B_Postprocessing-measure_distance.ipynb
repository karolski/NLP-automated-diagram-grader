```python
import pandas as pd
from Levenshtein import distance

MIN_THRESHOLD_OF_SIMILARITY = 0.2
```

```python
file_name = "avg_wv2_matched.csv"
results = pd.read_csv(file_name)
results = results[results["AVG-W2V_similarity"]>MIN_THRESHOLD_OF_SIMILARITY]
results['aimed_field'] = results['AVG-W2V_aimed_field'].str.extract(r'(\d+)')
results['student_cat'] = results["IDStud"].map(str) +";"+ results["Category"]
results['student_cat_field'] = results['student_cat'].map(str) + results["Fieldname"].map(str)
results
```

```python
temp_frame = results.copy()
temp_frame.set_index('student_cat_field', inplace=True)
temp_frame
```

```python
words = {"category":[],"id":[],"order":[], "score_dist":[],"similarity_sum":[]}
aimed_field = temp_frame["aimed_field"]
similarity = temp_frame["AVG-W2V_similarity"]
for student_cat in results["student_cat"].drop_duplicates(): #not tested
    words["category"].append(student_cat.split(';')[1])
    words["id"].append(student_cat.split(";")[0])
    word = ""
    sum_similarity = 0
    for i in range(1,5):
        try: 
            word += str(aimed_field.loc[student_cat+str(i)])
            sum_similarity += similarity.loc[student_cat+str(i)]
        except KeyError: word += ""        
    words["order"].append(word)
    ldistance = 5-distance(word, "1234")
    words["score_dist"].append(ldistance)
    words["similarity_sum"].append(sum_similarity)
words_frame = pd.DataFrame(words)
words_frame.head()
```

```python
self_assesment = pd.read_csv("self_assesment.csv", sep=";")
text_numbers = {1:"Beton", 2:"Geld", 3:"Suez", 4:"Muziek", 5:"Botox", 6:"Metro"}
self_assesment["category"] = self_assesment["TextNumber"].apply(lambda n: text_numbers[n])
self_assesment["id"] = self_assesment["IDStudent"]+self_assesment["IDClass"]
self_assesment.head()
```

```python
result_frame=words_frame.merge(self_assesment, on=["id", "category"])
result_frame.head()
```

```python
correlations = result_frame[["score_dist","similarity_sum","D3JOLF","D3JOLR"]].corr()
correlations
```

```python
results_filename = "logical_order_score_similarity>"+str(MIN_THRESHOLD_OF_SIMILARITY)+".csv"
correlations_filename = "correlations_similarity>"+str(MIN_THRESHOLD_OF_SIMILARITY)+".csv"
result_frame.to_csv(results_filename, index=False)
correlations.to_csv(correlations_filename)
```
